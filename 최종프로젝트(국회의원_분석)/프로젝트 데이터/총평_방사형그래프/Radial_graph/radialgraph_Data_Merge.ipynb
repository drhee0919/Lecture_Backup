### 기타 자료들 합치기

```python
# 1) 긍부정 분석 점수
PN_score = pd.read_csv('./data/parameters/PN_score.csv',
                             encoding='utf-8',
                             index_col=0)

# 2) 공약이행도 점수 
manifesto_score = pd.read_excel('./data/parameters/manifesto_score.xlsx',
                                sheet_name=1,
                                encoding='utf-8',
                                index_col=0)

# 3) 소통 일치도 점수 
com_score = pd.read_csv('./data/parameters/com_score.csv',
                        encoding='utf-8',
                        index_col=0)

# 4) 대표의안개수 점수
bill_score = pd.read_csv('./data/parameters/bill_score.csv',
                         encoding='utf-8',
                         index_col=0)

# 5) 회의록 점수
meet_score = pd.read_csv('./data/parameters/meet_score.csv',
                         encoding='utf-8',
                         index_col=0)


display(PN_score)
display(manifesto_score)
display(com_score)
display(bill_score)
display(meet_score)
```

```python
# 회의록 점수 평균값이 없으므로 만든다. 
meet_score["mean_score"] = meet_score["meet_score"].mean()
display(meet_score)
```

```python
# 대표의안의 인덱스가 이름으로 되어있어 나머지와 일치화 시킨다
bill_score.reset_index(inplace=True)
display(bill_score)
```

```python
# 공약이행률도 마찬가지로 처리한다 
manifesto_score.reset_index(inplace=True)
display(manifesto_score)
```

```python
#내용들을 붙여보자 (역순으로 회의록, 대표의안, 소통일치도, 공약이행도, 긍부정분석)

#1) 회의록
new_df = meet_score[['name','meet_score','mean_score']]
display(new_df)

#2) 회의록 + 대표의안
new_df = pd.concat([new_df, bill_score[['bill_score','bill_score_mean']]],
                    axis=1)
                    
display(new_df)           

#3) 회의록+대표의안+소통일치도
new_df = pd.concat([new_df, com_score[['com_score', 'avg']]],
                    axis=1)
display(new_df)

#4) 회의록+대표의안+소통일치도+공약이행도
new_df = pd.concat([new_df, manifesto_score[['점수','평균']]],
                   axis=1)
display(new_df)

#5) 회의록+대표의안+소통일치도+공약이행도+긍부정분석
new_df = pd.concat([new_df, PN_score[['긍정비율','평균']]],
                   axis=1)
display(new_df)
```

```python
# 컬럼 이름을 새로 붙여보자 
new_df.columns = ["이름", 
                  "회의", "회의_평균",
                  "의안","의안_평균",
                  "소통일치도","소통일치도_평균",
                  "공약이행도","공약이행도_평균",
                  "긍부정비율","긍부정비율_평균"]
display(new_df)
```

```python
# 파일로 저장하자 
new_df.to_csv('./data/merged_parameters.csv', encoding='utf-8')
```
